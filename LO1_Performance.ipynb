# LO1 Performance

## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import time
import plotly.express as px
```

## Load Dataset

```python
dists = [np.random.normal(size=n) for n in (np.array(range(21)) * 1_000_000)]
```

```python
execution_times = []

for dist in dists:
    start_time = time.time()

    px.histogram(dist)

    end_time = time.time()

    if len(dist) > 1:
        execution_times.append({'n': len(dist), 'time': end_time-start_time})

times = pd.DataFrame(execution_times)
```

```python
times.plot(x='n', y='time')
plt.legend().set_visible(False)
plt.xlabel('Number of samples')
plt.ylabel('Time to execute')
plt.title('Time to execute plot for each number of samples')
plt.show()
```
