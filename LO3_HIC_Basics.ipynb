## Import

```python
from pathlib import Path
from joblib import load

import plotly.express as px
```

### Import Dataframe

```python
source_path = Path('./data/clean/employees.dump')
df = load(source_path)

df.head()
```

```python
px.histogram(df,
             x='direktion_name',
             y='wert',
             color='indikator',
             animation_frame='jahr',
             barmode='group',
             labels={
                 'direktion_name': 'Name der Direktion',
                 'wert': 'Number of employees',
                 'indikator': 'Indikator'
             },
             range_y=(0, 20_000),
             title='Entwicklung der Anzahl Mitarbeiter pro Direktion')
```
